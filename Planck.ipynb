```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import h, c, k

def frecuencia_planck(Bv, T):
    return (2 * h * Bv**3 / c**2) * (1 / (np.exp((h * Bv) / (k * T)) - 1))


T = 5700  


Bv = np.logspace(5, 15, 200)  # Desde 10^5 Hz hasta 10^15 Hz


R_Bv = frecuencia_planck(Bv, T)

plt.figure(figsize=(10, 8))
plt.loglog(Bv, R_Bv, label=f'T = {T} K')
plt.title('Ecuacin de Planck a 5700 K en funcin de la frecuencia')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Radiancia Espectral (WsrmHz)')
plt.grid(True, which="both", ls=":")
plt.legend()
plt.show()
```
